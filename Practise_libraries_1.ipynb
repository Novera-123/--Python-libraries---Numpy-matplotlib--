```python
#Combined NumPy and Matplotlib:
#Data Manipulation and Visualization:

#Create a NumPy array with values from 0 to 100 in steps of 5.
#Plot a bar graph of the squared values of the array.

X=np.arange(0,101,5)
print(X)
values_square=np.square(X)
plt.bar(X,values_square, width =4)
plt.xlabel('Values')
plt.ylabel('Squared values')
plt.title('Bar graph of square values')
plt.show()
```

```python
#Trigonometric Functions:

#Generate an array of angles (in radians) from 0 to 2 with a step size of /4.
#Plot the sine and cosine functions in the same figure.

angles = np.arange(0, np.pi+np.pi/4, np.pi/4)
sine_values = np.sin(angles)
cosine_values = np.cos(angles)

plt.plot(angles,sine_values, label ='Sin')
plt.plot(angles,cosine_values,label = 'cos')

plt.xlabel('Angle(Radians)')
plt.ylabel("Function value")
plt.title("Sine and cosine function")

plt.legend()
plt.show()
```

```python
#Question 1: Array Operations and Visualization
#Array Operations:

#Create a NumPy array with values from 0 to 10 in steps of 0.5.
#Calculate the square root of each element in the array.
#Visualization:

#Plot the original array against the squared root values using Matplotlib.
#Customize the plot with labels and a title.

new_array = np.arange(0,10,0.5)
Sq_array = np.sqrt(new_array)
plt.plot(new_array,Sq_array, label="Original value",marker='o')
plt.plot(Sq_array,Sq_array, label ="Squared value",marker='s')




plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Original Values vs. Square Root Values")

plt.legend()
plt.show()
```

```python
#Array Manipulation:

#Create a 2D NumPy array with dimensions 3x4, filled with random integers.
#Calculate the sum and mean of each column.
#Visualization:

#Plot a histogram of the flattened values of the 2D array.
#Customize the plot with labels and a title.

new_array = np.random.randint(0,10, size = (3,4))
print(new_array)
sum_array = np.sum(new_array, axis=0)
mean_array = np.mean(new_array, axis =0)
flatten_values = new_array.flatten()
print(flatten_values)
plt.hist(flatten_values, bins ='auto',color='red', alpha = 0.8,rwidth=0.9)

plt.xlabel=("Values")
plt.ylabel=("Frequency")
plt.title = ("Histogram of flattened values")

plt.show()
```
